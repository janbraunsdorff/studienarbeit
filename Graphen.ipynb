```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (16, 5)
```

```python
def print_plot(acc, loss, c):
    f = plt.figure()
    ax = f.add_subplot(1, 3, 1)
    plt.plot(acc)
    plt.xlabel('Epochen')
    plt.ylabel('Genauigkeit')
    plt.title('Fehler im Trainigsdaten')
    ax.set_ylim(0, 1000)


    ax = f.add_subplot(1, 3, 2)
    plt.plot(loss)
    plt.xlabel('Epochen')
    plt.ylabel('Fehler')
    plt.title('Durchschnittlicher Fehler in Monaten')
    ax.set_ylim(0, 30)

    ax = f.add_subplot(1, 3, 3)
    for i in c:
        plt.plot(i)
    plt.xlabel('Epochen')
    plt.ylabel('Richtiges Alter %')
    plt.title('Fehlerklassen')
    plt.legend(['C4', 'C12', 'C24'])
    ax.set_ylim(0, 100)

    plt.show()
```
