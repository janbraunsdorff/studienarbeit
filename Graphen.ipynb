```python
def print_plot(acc, loss, c):
    f = plt.figure()
    ax = f.add_subplot(1, 3, 1)
    plt.plot(acc)
    plt.xlabel('Epochen')
    plt.ylabel('Genauigkeit')
    plt.title('Fehler im Trainigsdaten')
    ax.set_ylim(0, 250)


    ax = f.add_subplot(1, 3, 2)
    plt.plot(loss)
    plt.xlabel('Epochen')
    plt.ylabel('Fehler')
    plt.title('Durchschnittlicher Fehler in Monaten')
    ax.set_ylim(0, 15)

    ax = f.add_subplot(1, 3, 3)
    for i in c:
        plt.plot(i)
    plt.xlabel('Epochen')
    plt.ylabel('Richtiges Alter %')
    plt.title('Fehlerklassen')
    plt.legend(['C4', 'C12', 'C24'])
    ax.set_ylim(0, 100)

    plt.show()
```

```python
def binar(x):
    return 0 if x < 0 else 1
def lin(x):
    return x
def sig(x):
    return 1 / (1 + math.exp(-x))

def relu(x):
    return 0 if x < 0 else x


def Lrelu(x):
    return x * 0.1 if x < 0 else x

def Swish(x):
    return x  * sig(x)
def Prelu(x):
    return x * 0.2 if x < 0 else x

def Erelu(z):
    return z if z >= 0 else 0.2*(math.pow(math.e, z) - 1)
```
